```python
from utils import image_to_text
import spacy
from matplotlib import pyplot as plt
from model import Net
import torch
from dataset import TestDataset
from torch.utils.data import DataLoader
import torch.optim as optim
from tqdm import tqdm
import torch.nn.functional as F
from utils import pred_to_dict
from utils import extract_date
```

```python
# Defining data and model
batch_size = 16
test_dataset    = TestDataset()
test_dataloader = DataLoader(test_dataset, batch_size = 16, shuffle=True)
device = torch.device('cuda:0')
model = Net(len(test_dataset.VOCAB)+1).to(device)
model.load_state_dict(torch.load("models/best-model-parameters.pt"))
model.eval()
```

```python
answers = {}
# Running
with torch.no_grad():
    loop = tqdm(test_dataloader, leave = True)
    for batch in loop:
        input_ids   = batch[0].to(device)
        output      = model(input_ids)
        probs       = F.softmax(output, dim = 2)
        prob, pred  = torch.max(probs, dim = 2)
        prob        = prob.squeeze().cpu().numpy()
        pred        = pred.squeeze().cpu().numpy()
        for i in range(batch_size):
            text = batch[1][i]
            name = batch[2][i]
            dic  = pred_to_dict(text, pred[i], prob[i])
            dic['date'] = extract_date(text)
            answers[name] = dic
```

```python
import json
for filename, dictionary in answers.items():
    with open("models/answers/1.0/"+filename+".txt", "w") as outfile:
        json.dump(dictionary, outfile)
```
